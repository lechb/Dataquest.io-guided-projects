```python
import pandas as pd
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head(5))
print(recent_grads.tail(5))
print(recent_grads.describe())
```

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print(raw_data_count)
print(cleaned_data_count)
```

```python
recent_grads.plot(x='Sample_size',y='Median',kind='scatter',title='Median vs sample')
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',title='Unemployed vs sample')
recent_grads.plot(x='Full_time',y='Median',kind='scatter',title='Median vs Full_time')
recent_grads.plot(x='ShareWomen',y='Unemployment_rate',kind='scatter',title='Unemployed vs Share of women')
recent_grads.plot(x='Men',y='Median',kind='scatter',title='Median vs men')
recent_grads.plot(x='Women',y='Median',kind='scatter',title='Median vs Women')
```

Use the plots to explore the following questions:

- Do students in more popular majors make more money?
More popular majors does not guarantee making more money - less popular majors have higher probability that students could make more money
- Do students that majored in subjects that were majority female make more money?
Student where majority was female earn less money than those if men were majority. 
- Is there any link between the number of full-time employees and median salary?
There is not high correlation between full-time employment and median salary. It is possible that majors having lower number of full-time employees earn more money but it could be connected with number of students which charts above don't show us.
